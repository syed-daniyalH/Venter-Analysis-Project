```python
import pandas as pd
from pathlib import Path
workbook_path = Path("Venter_Analysis.xlsx")
if not workbook_path.exists():
    alt_path = Path("Vender Analysis") / "Venter_Analysis.xlsx"
    if alt_path.exists():
        workbook_path = alt_path
collections_df  = pd.read_excel(workbook_path, sheet_name=["Collections"])
items_df  = pd.read_excel(workbook_path, sheet_name=["Items"])
collections_df = collections_df["Collections"]
items_df = items_df["Items"]

print("\nItems Columns:", items_df.columns.tolist())

print("\nCollections Columns:", collections_df.columns.tolist())
```

```python

collections_df = collections_df.drop(
    columns=["Color Code", "Size", "Unit Level", "Code", "Barcode","Picture","Is Active"],
    errors='ignore'
)

```

```python
collections_df
```

```python
# Combine Name + Brand
items_df["Name"] = items_df["Name"].str.cat(items_df["Brand"], sep=" - ")

# Combine Secondary Name + Secondary Brand
items_df["Secondary Name"] = items_df["Secondary Name"].str.cat(items_df["Secondary Brand"], sep=" - ")

items_df
```

```python
items_df = items_df.drop(
    columns=[
        "Brand", "Secondary Brand"
    ],errors='ignore')
items_df
```

```python
collections_df['Secondary Item Name']=collections_df['Secondary Item Name'].str.cat(collections_df['Secondary Color'])
collections_df['Item Name']=collections_df['Item Name'].str.cat(collections_df['Color'])
collections_df
```

```python
collections_df = collections_df.drop(
    columns=[
        "Color", "Secondary Color"
    ],errors='ignore')
collections_df
```

```python
# Conversion rate
conversion_rate = 0.2153927  

if 'Price' in collections_df.columns:
    collections_df['Price_IQD'] = (
        collections_df['Price']
        .astype(str)
        .str.replace("د.ع", "", regex=False)
        .str.replace(",", "", regex=False)
        .str.strip()
        .replace({"": "0", "00": "0", "01": "1"})
        .astype(float)
    )
    collections_df['Price_PKR'] = collections_df['Price_IQD'] * conversion_rate

# Preview
print("Items DF:\n", items_df.head(), "\n")
print("Collections DF:\n", collections_df.head(), "\n")
```

```python
output_dir = Path("outputs")
output_dir.mkdir(parents=True, exist_ok=True)
output_file = output_dir / "Venter_Analysis_processed.xlsx"

with pd.ExcelWriter(output_file, engine="openpyxl") as writer:
    items_df.to_excel(writer, sheet_name="Items", index=False)
    collections_df.to_excel(writer, sheet_name="Collections", index=False)

print("Data saved successfully to", output_file)
items_df.head()
```

```python
items_df["Sheet"] = "Items"
collections_df["Sheet"] = "Collections"

merged_df = pd.concat([items_df, collections_df], ignore_index=True)

merged_count = len(merged_df)
print(f"Total merged rows: {merged_count}")
```

```python

matched_df = collections_df[collections_df["Item Id"].isin(items_df["Id"])]

matched_df = matched_df[["Item Id", "Item Name", "Secondary Item Name","Price_IQD", "Price_PKR"]]

print("From Collections Sheet: \n",matched_df)
```
